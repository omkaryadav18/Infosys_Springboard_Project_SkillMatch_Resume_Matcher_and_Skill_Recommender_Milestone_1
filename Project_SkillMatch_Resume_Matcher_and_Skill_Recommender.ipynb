# **MATPLOTLIB**

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[2,4,1])
plt.show()
```

```python
x=np.arange(1,11)
y=2*x
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.show()
```

```python
y1=2*x
y2=3*x
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='--',linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
plt.scatter(x,a,marker="*",c='black',s=100)
plt.show()
```

```python
student = {'bob':87,'matt':56,'sam':97}
names=list(student.keys())
values=list(student.values())
plt.bar(names,values)
plt.show()
```

Line Plot

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.plot(x, y1,color='g',linestyle=':',linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=2)
plt.show()
```

**Box Plot**

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]

data = list([one,two,three])
plt.boxplot(data)
plt.show()
```

```python
plt.violinplot(data,showmedians=True)
plt.show()
```

**Pie Chart**

```python
fruit=['apple','orange','mango','guava']
quantity=[67,34,100,29]
plt.pie(quantity,labels=fruit,colors=['r','orange','y','g'])
plt.show()
```

**Doughnut Chart**

```python
plt.pie(quantity,labels=fruit,radius=2,colors=['r','orange','y','g'])
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
